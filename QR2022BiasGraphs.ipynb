```python
import matplotlib.pyplot as plt
import numpy as np 
import scipy 
import math
```

```python
def likelihood(Heads, Tails):
    Bias = np.linspace(0, 1, 1000)
    Likelihood = Bias ** Heads * (1 - Bias) ** Tails / scipy.special.comb(Heads + Tails, Heads)

    plt.plot(Bias, Likelihood)
    plt.title("Likelihood of Bias")
    # naming the x axis
    plt.xlabel('B')
    # naming the y axis
    plt.ylabel('Likelihood')
    plt.show()
```

```python
Heads = 50
Tails = 50
likelihood(Heads, Tails)
```
